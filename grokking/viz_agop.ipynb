```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
root = '/scratch/bbjr/mallina1/grokking_output/nov21-grokking/t1lywd7w'

agop = np.load(os.path.join(root, 'agop.npy'))
```

```python
plt.imshow(agop, cmap='gray')
plt.colorbar()
```

```python
vals, vecs = np.linalg.eig(agop)
```

```python
print(vals[:10])
```

```python
n_comps = 10
vec1 = vecs[:,0:n_comps]
agop_reconstruction = vec1 @ np.diag(vals[0:n_comps]) @ vec1.T
plt.imshow(agop_reconstruction, cmap='gray')
```

```python
dft2 = np.zeros((19, 19)).astype(np.complex128)
offset1 = np.pi/4

rads = np.arange(1, 10)
offset2 = np.pi/rads
for rad in rads:
    n_points = 19
    lo = 0
    hi = 2*np.pi
    interval = (hi - lo)/n_points
    
    xs = np.concatenate((rad*np.cos(offset2[rad-1]+np.arange(0, 2*np.pi, interval)), rad*np.cos(offset2[rad-1]+offset1 + np.arange(0, 2*np.pi, interval))))
    ys = np.concatenate((rad*np.sin(offset2[rad-1]+np.arange(0, 2*np.pi, interval)), rad*np.sin(offset2[rad-1]+offset1 + np.arange(0, 2*np.pi, interval))))

    xs1 = [xs[i] for i in range(0, int(len(xs)/2))]
    ys1 = [ys[i] for i in range(0, int(len(ys)/2))]
    xs2 = [xs[i] for i in range(int(len(xs)/2), len(xs))]
    ys2 = [ys[i] for i in range(int(len(ys)/2), len(ys))]
    plt.scatter(xs2, ys2)
    break

    dft2[:,rad] = (np.ones(len(xs1))*1j*xs1) + ys1
    dft2[:,rad+len(rads)] = np.ones(len(xs1))*1j*xs2 + ys2
# dft2[:,0] = np.sqrt(19)/2
```

```python
real = np.real(dft2)
comp = np.imag(dft2)
plt.scatter(comp, real, s=2)
```

```python
circ2 = np.fft.ifft(dft2)
```
